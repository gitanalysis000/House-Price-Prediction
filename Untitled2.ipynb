```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("housing.csv")
```

```python
data 
```

```python
#median_nouse_value is our target value as we are predicting it and the rest values are used as input values i.e used for predicting the median house value
#we need to preprocess ocean proximity as we can't feed text variables into the model and we need to scale the other values.
data.info() #for checking the null values.
```

```python
data.dropna(inplace=True) # nan (not a number) values are removed 
```

```python
data.info()
```

```python
#data preprocessing
from sklearn.model_selection import train_test_split

X = data.drop(['median_house_value'], axis=1) #full data frame withou the column median house value
y = data['median_house_value'] 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2) #test is used after training and hyperparameter tuning  
```

```python
train_data = X_train.join(y_train)
```

```python
train_data 
```

```python
train_data.hist(figsize=(12, 6)) #corelation with target variable , distributuin of various variables
```

```python
plt.figure(figsize=(12, 6))
sns.heatmap(train_data.corr(numeric_only = True), annot=True, cmap="YlGnBu") #heatmap of correlation matrix -> train_data.corr()
```

```python
#Here we can see that the median income value highly corelates with the median house value
#Herew we can see some value are skewed (not in gausian bell sense) so we have to take the logarithm of those features and check their distributions
train_data['total_rooms'] = np.log(train_data['total_rooms'] + 1)
train_data['total_bedrooms'] = np.log(train_data['total_bedrooms'] + 1)
train_data['population'] = np.log(train_data['population'] + 1)
train_data['households'] = np.log(train_data['households'] + 1)
```

```python
train_data.hist(figsize = (12,6))
```

```python
train_data.ocean_proximity.value_counts()
#hot encoding 
train_data = train_data.join(pd.get_dummies(train_data.ocean_proximity)).drop(['ocean_proximity'], axis=1)
```

```python
plt.figure(figsize=(12, 6))
sns.heatmap(train_data.corr(numeric_only = True), annot=True, cmap="YlGnBu")
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x = "latitude", y ="longitude", data = train_data, hue="median_house_value", palette="coolwarm") #more red = more expensive , more blue = more cheaper
```

```python
#feature engineering merging few features into a new feature to reduce them(reduce the siumilar result parameters)
train_data['bedroom_ratio'] = train_data['total_bedrooms'] / train_data['total_rooms']
train_data['household_rooms'] = train_data['total_rooms'] / train_data['households']
```

```python
plt.figure(figsize=(12, 6))
sns.heatmap(train_data.corr(numeric_only = True), annot=True, cmap="YlGnBu")
```

```python
#linear Regression
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler 
scaler = StandardScaler()
X_train, y_train = train_data.drop(['median_house_value'] , axis = 1) , train_data['median_house_value']
X_train_s = scaler.fit_transform(X_train)
reg = LinearRegression()
reg.fit(X_train_s, y_train) #this model doesnot require hyperparameter tuning
```

```python
test_data = X_test.join(y_test)

test_data['total_rooms'] = np.log(test_data['total_rooms'] + 1)
test_data['total_bedrooms'] = np.log(test_data['total_bedrooms'] + 1)
test_data['population'] = np.log(test_data['population'] + 1)
test_data['households'] = np.log(test_data['households'] + 1)

test_data = test_data.join(pd.get_dummies(test_data.ocean_proximity)).drop(['ocean_proximity'], axis=1)

test_data['bedroom_ratio'] = test_data['total_bedrooms'] / test_data['total_rooms']
test_data['household_rooms'] = test_data['total_rooms'] / test_data['households']

X_test, y_test = test_data.drop(['median_house_value'] , axis = 1) , test_data['median_house_value'] #for eliminating the flot value error for few features 
```

```python
X_test_s = scaler.fit_transform(X_test)
```

```python
reg.score(X_test_s, y_test) #the result is not too bad but also not too good so we need to use a better model for better score.
#To get better score do use scaler from sklearn.preprocessing import StandardScaler scaler = StandardScaler() in new cell X_train_s = scaler.fit_transform(X_train) change input of reg.fit too 
```

```python
#RFR
from sklearn.ensemble import RandomForestRegressor #ensembele is used to combine different models in this case descision trees

forest = RandomForestRegressor()

forest.fit(X_train_s, y_train)
```

```python
forest.score(X_test_s, y_test)
```

```python
#we are using cross validation i.e the data is split into k folds we use k-1 for training and 1 for testing(evaluation)
from sklearn.model_selection import GridSearchCV #cross validation

forest = RandomForestRegressor()
#parameter grid cell ,below grid it will go through the all combinations, we need to have a postive score
param_grid = {
    "n_estimators": [100,200,300],
    "max_features": [2,4,6,8],
    #"min_samples_split": [2,4,6,8],
    #"max_depth": [0,4,8],
}

grid_search = GridSearchCV(forest, param_grid, cv=5,
                          scoring = "neg_mean_squared_error",
                          return_train_score=True)

grid_search.fit(X_train_s,y_train) #this gives us the optimum model - best estimator/regressor
```

```python
best_forest = grid_search.best_estimator_
```

```python
best_forest.score(X_test_s, y_test) #usually you get better results if not then increase the estimators and max features and add
#We need to go to more the direction of the optimal value of the grid(check the website for their optiml values)
```
